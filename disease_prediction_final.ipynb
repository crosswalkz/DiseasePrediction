```python
#  User input
# --------------

import nltk
#nltk.download('stopwords')
from nltk.corpus import stopwords
#nltk.download('punkt')
en_stops = set(stopwords.words('english'))

unwanted_char = [',','.','/','?','!',':',';','"','(',')']
t = input()
for i in unwanted_char:
    t = t.replace(i,' ')
text = list(map(str,t.strip().split()))
print(text)

text_input = []
for word in text: 
    if word not in en_stops:
        text_input.append(word)

print('text_input : ',text_input)
```

```python

# Find cosine sim


def word2vec(word):
    from collections import Counter
    from math import sqrt

    # count the characters in word
    cw = Counter(word)
    # precomputes a set of the different characters
    sw = set(cw)
    # precomputes the "length" of the word vector
    lw = sqrt(sum(c*c for c in cw.values()))

    # return a tuple
    return cw, sw, lw

def cosdis(v1, v2):
    # which characters are common to the two words?
    common = v1[1].intersection(v2[1])
    # by definition of cosine distance we have
    return sum(v1[0][ch]*v2[0][ch] for ch in common)/v1[2]/v2[2]

# text_input = list(map(str,input("ENTER: ").split(',')))
sym = []
for i in text_input:
    for j in ds_s2['Symptom']:
        list_A = [i]             # User Input Symptoms
        list_B = [j]             # Symptoms list from dataset

        threshold = 0.75     # if needed
        
        for key in list_A: 
            for word in list_B:
                print('word: ',key)
                try:
                    # print(key)
                    # print(word)
                    res = cosdis(word2vec(word), word2vec(key))
                    # print(res)
                    print("The cosine similarity between : {} and : {} is: {}".format(word, key, res*100))
                    if res > threshold:
                        print('----------------')
                        print("Found a word with cosine distance > 75 : {} with original word: {}".format(word, key))
                        print('--------------\n')
                        sym.append(word)
                        
                except IndexError:
                    pass

print("\n\nSym : ",sym)
```

```python
a = np.array(ds_s2['Symptom'])
b = np.array(ds_s2['weight'])
# print(a)
# print('----------------')
# print(b)
```

```python
for i in range(len(sym)):
    for j in range(len(ds_s2['Symptom'])):
        if sym[i] == a[j]:
            sym[i] = b[j]
            print(sym[i],end=' ')


psym = sym
for i in range(17-len(psym)):
    psym.append(0)

print('\n\npsym : ',psym)
```

```python
model = DecisionTreeClassifier()
model.fit(data,lables)
prediction = model.predict([psym])
prediction
```

## Accuracy

```python
X_train, X_test, y_train, y_test = train_test_split(data, lables, shuffle=True, test_size=0.2)
model.fit(X_train, y_train)
predict = model.predict(X_test)
score = accuracy_score(y_test,predict)
score
```
